# Define function for plot

```python
def draw_NEMO_grid(lon, lat, color='k', linewidth=0.5):
    '''
        Simply draw NEMO tripolar grid on steorographic projection
    '''
    #fig
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
    #left
    proj=Basemap(projection='npstere', resolution='l', \
                    boundinglat=0, lon_0=90, round=True, ax=axes[0])
    proj.drawcoastlines(linewidth=1.5, linestyle='-', color='k', zorder=3)
    proj.drawlsmask(land_color=[0.5, 0.5, 0.5], ocean_color='None', lsmask=None, zorder=2)
    x, y=proj(lon, lat)
    proj.plot(x[:, :].T, y[:, :].T, color=color, linewidth=linewidth)
    proj.plot(x[:, :], y[:, :], color=color, linewidth=linewidth)
    #right
    proj=Basemap(projection='spstere', resolution='l', \
                    boundinglat=0, lon_0=90, round=True, ax=axes[1])
    ax=plt.gca()
    proj.drawcoastlines(linewidth=1.5, linestyle='-', color='k', zorder=3)
    proj.drawlsmask(land_color=[0.5, 0.5, 0.5], ocean_color='None', lsmask=None, zorder=2)
    x, y=proj(lon, lat)
    proj.plot(x[:, :].T, y[:, :].T, color=color, linewidth=linewidth)
    proj.plot(x[:, :], y[:, :], color=color, linewidth=linewidth)
    return fig, axes
```

```python
def draw_ORCA2_Arctic(lon, lat, lonlim, latlim, var, clev, var_name, CMap):
    '''
        Draw NEMO forcing data on Arctic
    '''
    fig=plt.figure(figsize=(14, 14))
    proj=Basemap(projection='npstere', resolution='l', \
                boundinglat=0, lon_0=90, round=True)
    ax=plt.gca()
    parallels=np.arange(-90, 90, 15)
    meridians=np.arange(50, 360, 60)
    proj.drawparallels(parallels, labels=[1, 1, 1, 1],\
                      fontsize=10, latmax=90)
    proj.drawmeridians(meridians, labels=[1, 1, 1, 1],\
                      fontsize=10, latmax=90)
    proj.drawcoastlines(linewidth=1.5, linestyle='-', color='k', zorder=3)
    proj.drawlsmask(land_color=[0.5, 0.5, 0.5], ocean_color='None', lsmask=None, zorder=2)
    x, y=proj(lon, lat)
    CS=proj.contourf(x, y, var, clevs, cmap=CMap, extend='both') # RdBu_r
    proj.contour(x, y, var, clevs, colors='k', linewidths=2.0)
    CBar=proj.colorbar(CS, location='right', size='5%', pad='10%')
    return fig, ax, proj
```

```python
def draw_ORCA2_global(lon, lat, lonlim, latlim, var, clev, var_name, CMap):
    '''
        Draw NEMO forcing data on a global scale
    '''
    fig=plt.figure(figsize=(14, 14))
    proj=Basemap(projection='cyl', llcrnrlat=latlim[0], urcrnrlat=latlim[1], \
            llcrnrlon=lonlim[0], urcrnrlon=lonlim[1], resolution='l')
    ax=plt.gca()
    parallels=np.arange(-90, 90, 60)
    meridians=np.arange(50, 360, 120)
    proj.drawparallels(parallels, labels=[1, 1, 1, 1],\
                      fontsize=10, latmax=90)
    proj.drawmeridians(meridians, labels=[1, 1, 1, 1],\
                      fontsize=10, latmax=90)
    proj.drawcoastlines(linewidth=1.5, linestyle='-', color='k', zorder=3)
    proj.drawlsmask(land_color=[0.5, 0.5, 0.5], ocean_color='None', lsmask=None, zorder=2)
    x, y=proj(lon, lat)
    CS=proj.contourf(x, y, var, clevs, cmap=CMap, extend='both') # RdBu_r
    proj.contour(x, y, var, clevs, colors='k', linewidths=2.0)
    CBar=proj.colorbar(CS, location='right', size='5%', pad='10%')
    return fig, ax, proj
```
