```python
#import copy
import glob
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt
from __future__ import division
from __future__ import print_function
from mpl_toolkits.basemap import cm
from mpl_toolkits.basemap import Basemap
from mpl_toolkits.basemap import shiftgrid
% matplotlib inline
```

```python
ptrc_filename=glob.glob('_data/MY_TRC/Six*1m*p*.nc')
print("found file: \n{}".format(ptrc_filename))
```

```python
ptrc_obj=nc.Dataset(ptrc_filename[0])
ptrc_obj.variables.keys()
latitude=ptrc_obj.variables['nav_lat'][:]
longitude=ptrc_obj.variables['nav_lon'][:]
```

```python
river_lon=[72, 79, 129, 162, -164, -133]
river_lat=[72, 73, 72.5, 70, 63, 70]
#lat_index = np.searchsorted(latitude.flat, river_lat, 'right')-1
#lat_i, lat_j = divmod(np.searchsorted(latitude.ravel(), river_lat), latitude.shape[1])
#lon_i, lon_j = divmod(np.searchsorted(longitude.ravel(), river_lon), longitude.shape[1])
#lon_index = np.searchsorted(longitude[0, :].flat, river_lon, 'right')-1
#river_grid=np.zeros(latitude.shape)
#river_grid[lat_index, lon_index]=1
lon_index=np.zeros([longitude.shape[0], len(river_lon)])
lat_index=np.zeros([len(river_lat), latitude.shape[1]])
for i in range(longitude.shape[0]):
    lon_index[i, :] = np.searchsorted(longitude[i, :].flat, river_lon, 'right')-1
for i in range(latitude.shape[1]):
    lat_index[:, i] = np.searchsorted(latitude[:, i].flat, river_lat, 'right')-1
    
lon_index.shape
#longitude[lon_index.astype(int)]
longitude[1, lon_index[1, :].astype(int)]
```

```python
fig=plt.figure(figsize=(7, 7))
proj=Basemap(projection='npstere', resolution='c', \
            boundinglat=50, lon_0=90, round=True)
ax=plt.gca()
## define parallels and meridians to draw.
parallels=np.arange(-90, 90, 15)
meridians=np.arange(0, 360, 60)
proj.drawparallels(parallels, labels=[1, 1, 1, 1],\
                  fontsize=10, latmax=90)
proj.drawmeridians(meridians, labels=[1, 1, 1, 1],\
                  fontsize=10, latmax=90)
proj.drawcoastlines(linewidth=2.0, linestyle='-', color='k')
# compute native x,y coordinates of grid.
x, y=proj(longitude, latitude)
#river_x, river_y=proj(river_lon, river_lat)
CS=proj.pcolor(x, y, river_grid, vmin=0, vmax=2, cmap=plt.cm.hot_r)
ax.set_title('Started points of rivers', fontsize=14, fontweight='bold', y=1.025)
plt.show()
```
