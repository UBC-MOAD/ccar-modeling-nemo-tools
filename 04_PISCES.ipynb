```python
import copy
import glob
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt
from __future__ import division
from __future__ import print_function
from mpl_toolkits.basemap import cm
from mpl_toolkits.basemap import Basemap
from mpl_toolkits.basemap import shiftgrid
% matplotlib inline
```

```python
ptrc_obj=glob.glob('_data/ORCA2_LIM_PISCES/*1y*ptrc_T.nc')
print("found file: \n{}".format(ptrc_obj))
diad_obj=glob.glob('_data/ORCA2_LIM_PISCES/*1y*diad_T.nc')
print("found file: \n{}".format(diad_obj))
```

```python
PTRC=nc.Dataset(ptrc_obj[0])
DIAD=nc.Dataset(diad_obj[0])
```

```python
depth=PTRC.variables['deptht'][:]
latitude=PTRC.variables['nav_lat'][:]
longitude=PTRC.variables['nav_lon'][:]
```

```python
DIC=PTRC.variables['DIC'][:][0, :, :, :] # mmol/m3
print('DIC: {} ~ {}'.format(np.min(DIC), np.max(DIC)))
TALK=PTRC.variables['Alkalini'][:][0, :, :, :]
print('TALK: {} ~ {}'.format(np.min(TALK), np.max(TALK)))
NO3=PTRC.variables['NO3'][:][0, :, :, :]
print('NO3: {} ~ {}'.format(np.min(NO3), np.max(NO3)))
NH4=PTRC.variables['NH4'][:][0, :, :, :]
print('NH4: {} ~ {}'.format(np.min(NH4), np.max(NH4)))
PO4=PTRC.variables['PO4'][:][0, :, :, :]
print('PO4: {} ~ {}'.format(np.min(PO4), np.max(PO4)))
```

```python
DIC_flux=DIAD.variables['Cflx'][0, :, :] # unit: mol/m2/s
print('DIC_flux: {} ~ {}'.format(np.min(DIC_flux), np.max(DIC_flux)))
pH=DIAD.variables['PH'][:][0, :, :, :]
print('pH: {} ~ {}'.format(np.nanmin(pH), np.nanmax(pH)))
```

```python
fig=plt.figure(figsize=(9, 9))
proj=Basemap(projection='npstere', resolution='c', \
            boundinglat=20, lon_0=90, round=True)
ax=plt.gca()
## define parallels and meridians to draw.
parallels=np.arange(-90, 90, 15)
meridians=np.arange(0, 360, 60)
proj.drawparallels(parallels, labels=[1, 1, 1, 1],\
                  fontsize=10, latmax=90)
proj.drawmeridians(meridians, labels=[1, 1, 1, 1],\
                  fontsize=10, latmax=90)
# draw coast & fill continents
proj.drawcoastlines(linewidth=2.0, linestyle='-', color='k')
# compute native x,y coordinates of grid.
x, y=proj(longitude, latitude)
proj.plot(x, y, 'k+')
ax.set_title('Modelling Grid (cpp_key "arctic")', fontsize=14, fontweight='bold', y=1.025)
plt.show()
```
